```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Gandhi\Machine Learning\date_range.csv")
df
```

```python
#Date column is missing so creating an object and specifying the range
range=pd.date_range(start="7/1/2019",end="7/30/2019",freq='B')
range
```

```python
#setting index
df.set_index(range,inplace=True)
df
```

```python
#Plotting 
%matplotlib inline
df.close.plot()
```

```python
#Selecting a particular range of dates
df["2019-07-01":"2019-07-10"]
```

```python
#getting the average value of the particular date range
df["2019-07-01":"2019-07-10"].close.mean()
```

```python
#Changing the frequency
df.asfreq('D',method='pad')
```

```python
#Creating a new date_range
range1=pd.date_range(start='1/1/2019',periods=50,freq='D')
range1
```
