```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("D:\\Gandhi\\Machine Learning\\replacing_missing_data.csv")
df
```

```python
#To replace the -99999 value with NaN and If the dateset has 2 special values, we can use list to replace those.
df_new=df.replace(['-99999','-88888'],np.NaN)
df_new
```

```python
# If we want to replace the event column also, we can make use of dictionary
df_new=df.replace({
    'temperature':['-99999','-88888'],
    'windspeed':['-99999','-88888'],
    'event':'0'
},np.NaN)
df_new
```

```python
#TO replace the alphabets from few columns
df_new=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)               #Regex are used to detect patterns
df_new
```

```python
#To replace a list with another list

#Creating a dataframe
df1=pd.DataFrame({
    'Name':['Manish','Gandhi','Anisha','Maryam'],
    'Score':['Poor','Bad','Good','Good']
})
df1
```

```python
df2=df1.replace(['Poor','Bad','Good','Good'],[0,1,2,3])
df2
```
