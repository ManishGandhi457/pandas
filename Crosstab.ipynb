```python
import pandas as pd
```

```python
df=pd.read_excel("D:\\Gandhi\\Machine Learning\\crosstab.xlsx")
df
```

```python
#it is used to modify the data into a x and y axis format. Here Nationality is x axis and Handedness is y  axis
pd.crosstab(df.Nationality,df.Handedness)   #pd,crosstab(df.x axis,df.y axis)
```

```python
pd.crosstab(df.Sex,df.Handedness,margins=True)   #margins is to get a total 
```

```python
#If i want to have 2 vvariables at column level
pd.crosstab(df.Sex,[df.Nationality,df.Handedness])
```

```python
#Using nationality as x axis
pd.crosstab([df.Nationality,df.Sex],df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness,margins=True)
```

```python
pd.crosstab(df.Sex,df.Handedness,normalize="index")
```
