```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Gandhi\Machine Learning\SHifting_Lagging.csv",parse_dates=True,index_col="Date")
df
```

```python
#It will shift the price column by 1 day
df.shift(1)
```

```python
#We can shift either to the right or to the left.. +ve value shifts value to the right and -ve value shifts value to the left
df.shift(-1)
```

```python
#Creating a previous day column to get to know more about shifting
df["Previous day price"]=df["Price"].shift(1)
df
```

```python
df["Price difference"]=df["Price"]-df["Previous day price"]
df
```

```python
#To find the percentage of price before 5 days
df["5 days percent"]=(df["Price"]-df["Price"].shift(5))*100/df["Price"].shift(5)
df
```

```python
#Reodering the dataframe as it was before
df=df["Price"]
df
```

```python
df.index
#Here frequency is none
```

```python
#Creating a date range with business frequency
df.index=pd.date_range(start="2017-08-15",periods=10,freq='B')
df.index
```
